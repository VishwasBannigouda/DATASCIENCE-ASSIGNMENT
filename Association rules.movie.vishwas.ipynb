```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
data=pd.read_csv('C:\\Users\\vishw\\excelr drive files\\Assignments questions of excelr\\Association_rules\\my_movies.csv')
data
```

```python
data.drop(['V1','V2','V3','V4','V5'],inplace=True,axis=1)
```

```python
data
```

# Apriori algorithm

```python
frequent_items=apriori(data,min_support=0.05,use_colnames=True)
frequent_items
```

```python
rules=association_rules(frequent_items,metric='lift',min_threshold=1)
rules
```

```python
rules[(rules['lift']>= 6) & (rules['confidence'] >= 0.8)]
```
