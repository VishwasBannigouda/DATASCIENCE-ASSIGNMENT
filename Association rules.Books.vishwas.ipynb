# Apriori algorithm

```python
books=apriori(data,min_support=0.005,use_colnames=True)
books
```

```python
rules=association_rules(books,metric="lift",min_threshold=1)
rules
```

```python
rules.sort_values('lift',ascending=False,)
```

```python
rules.sort_values('lift',ascending = False)
```
